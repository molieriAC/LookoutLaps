```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
laps_df = pd.read_csv('lookout_data.csv', usecols=['id', 'elapsed_time', 'start_date_local', 'average_watts', 'average_heartrate', 'max_heartrate'])
```

```python
laps_df.head()
```

```python
# sort by start_date_local
laps_df.sort_values(by='start_date_local', inplace=True)

laps_df['start_date_local'] = pd.to_datetime(laps_df['start_date_local'])

# keep only data from 2021 and later
laps_df2 = laps_df[pd.DatetimeIndex(laps_df['start_date_local']).year >= 2021].copy()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.lineplot(x='start_date_local', y='elapsed_time', data=laps_df2)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
